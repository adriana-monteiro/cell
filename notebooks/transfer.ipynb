#### 1.5.1 Function show_sample_batch(ds, grid_size=(1, 3))

````python
# To invoque: show_sample_batch(ds_train)

def show_sample_batch(ds, grid_size=(1, 3)):
  '''
  # def show_sample_batch(ds, grid_size=(3, 3)):
  # To invoque: show_sample_batch(ds_train)
  '''
  n_images = grid_size[0]*grid_size[1]
  # Get a batch via iteration
  iter_ = iter(ds)
  batch_x, batch_y = iter_.next()
  batch_x, batch_y = batch_x[0:n_images], batch_y[0:n_images]
  # Plot the images
  plt.figure(figsize=(10, 10))
  for i, (img, y) in enumerate(zip(batch_x, batch_y)):
      ax = plt.subplot(grid_size[0], grid_size[1],  i + 1)
      plt.imshow(tf.cast(img, tf.int32))
      plt.title("{} image of class \"{}\"".format(img.shape, y), size=8)
      plt.axis("off")
````

#### 1.5.2 Function plot_confusion_matrix

````python
# Function to plot confusion matrices 

def plot_confusion_matrix(train_actual, train_predicted, val_actual, val_predicted, labels):
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))

    # Plot confusion matrix for training data
    train_cm = tf.math.confusion_matrix(train_actual, train_predicted)
    train_ax = sns.heatmap(train_cm, annot=True, fmt='g', ax=axes[0])
    train_ax.set_title('Training Confusion Matrix')
    train_ax.set_xlabel('Predicted Action')
    train_ax.set_ylabel('Actual Action')
    train_ax.set_xticklabels(labels, rotation=90)
    train_ax.set_yticklabels(labels, rotation=0)

    # Plot confusion matrix for validation data
    val_cm = tf.math.confusion_matrix(val_actual, val_predicted)
    val_ax = sns.heatmap(val_cm, annot=True, fmt='g', ax=axes[1])
    val_ax.set_title('Validation Confusion Matrix')
    val_ax.set_xlabel('Predicted Action')
    val_ax.set_ylabel('Actual Action')
    val_ax.set_xticklabels(labels, rotation=90)
    val_ax.set_yticklabels(labels, rotation=0)

    plt.tight_layout()
    plt.show()

````

### 1.5.3 Function for loading the dataset from the directory

````python
def load_from_directory(path_to_folder, image_size = (192,256), crop_to_aspect_ratio = True, color_mode = 'rgb', batch_size = 64, labels = 'inferred', label_mode='categorical', class_names=['Benign', '[Malignant] Pre-B','[Malignant] Pro-B', '[Malignant] early Pre-B'], shuffle = True, seed = 1847):
  ds = image_dataset_from_directory(path_to_folder,
                                          image_size = image_size,
                                          crop_to_aspect_ratio = crop_to_aspect_ratio,
                                          color_mode = color_mode,
                                          batch_size = batch_size,
                                          labels = labels,
                                          label_mode = label_mode,
                                          class_names = class_names,
                                          shuffle = shuffle,
                                          seed = seed
                                          )
  return ds
````

### 1.5.4 Function for loading datasets ready for training

````python
def load_datasets(returnValidationDs = False, balanced = False, blackVersions = False, returnTestDs = False, image_size = (192,256)):
  ds_train_balanced = None
  ds_train_black = None
  ds_train_balanced_black = None
  ds_val = None
  ds_test = None
  ds_train = load_from_directory(train_folder, image_size = image_size)

  if returnValidationDs == True:
    ds_val = load_from_directory(val_folder, image_size = image_size)
  
  if returnTestDs == True:
    ds_test = load_from_directory(test_folder, image_size = image_size)

  if balanced == True:
    ds_train_augmented = load_from_directory(dicParams["data_folder_preprocess_train_augmented"], image_size = image_size)
    ds_train_balanced = ds_train.concatenate(ds_train_augmented)
    ds_train_balanced = ds_train_balanced.shuffle(10000, reshuffle_each_iteration=True)
  
  if blackVersions == True:
    ds_train_black = load_from_directory(dicParams["data_folder_preprocess_train_black"])

  if balanced == True and blackVersions == True:
    ds_train_balanced_black = load_from_directory(dicParams["data_folder_preprocess_train_augmented_black"], image_size = image_size)
    ds_train_balanced_black = ds_train_black.concatenate(ds_train_augmented).concatenate(ds_train_balanced_black)
    ds_train_balanced_black = ds_train_balanced_black.shuffle(10000, reshuffle_each_iteration=True)
        
  return ds_train, ds_val, ds_test, ds_train_balanced, ds_train_black, ds_train_balanced_black
````

````python
def count_per_cat(ds):
  labelsPerCat = {0: 0, 1: 0, 2: 0, 3: 0}
  for images_batch, labels_batch in ds:
    for label in labels_batch:
      label_index = np.argmax(label, axis=-1)
      labelsPerCat[label_index] = labelsPerCat[label_index] + 1
  return labelsPerCat
````

## 3. Transfer Learning Model

### 3.1 Transfer Learning using ResNet50V2

In the cell below, team define a convolutional neural network (CNN) for image classification using transfer learning. The ResNet50V2 pre-trained model is used as the convolutional base, which has been trained on the large ImageNet dataset, and additional layers are added on top for multiclass classification of a new set of images. At the moment of base_model creation team pass three parameters. The *include_top=False* parameter specifies that the top (classification) layer of the pre-trained model should not be included. The *weights='imagenet'* parameter includes the pre-trained weights for the ResNet50V2 model that were trained on the ImageNet dataset. The *input_shape* of the model is set to be a 3-channel (RGB) image with dimensions 224x224, which is required for ResNet50V2.

For the first iteration, it is important to freeze the convolutional base before you compile and train the model. Freezing prevents the weights in a given layer from being updated during training.

To adapt the base_model for the classification of 4 classes, team added Average Pooling layer, Batch Normalization layer, Dropout layer and final Dense layer with required dimensionality and softmax activation for multiclass classification.


````python
base_model = keras.applications.ResNet50V2(
                        include_top=False, # Exclude ImageNet classifier at the top
                        weights='imagenet',
                        input_shape=(dicParams['image_size'], dicParams['image_size'], 3)
                        )

base_model.trainable = False

inputs = keras.Input(shape=(dicParams['image_size'], dicParams['image_size'], 3))
x = keras.applications.resnet_v2.preprocess_input(inputs)
x = base_model(x, training=False)

# Rebuild top layers
x = layers.GlobalAveragePooling2D()(x) # Average pooling operation
x = layers.BatchNormalization()(x) # Introduce batch norm
x = layers.Dropout(0.2)(x)  # Regularize with dropout layer to prevent overfitting

# Flattening to final layer - Dense classifier with 4 units (multi-class classification)
num_classes = 4
outputs = layers.Dense(num_classes, activation='softmax')(x)
model = keras.Model(inputs, outputs)
````

The *ResNet50V2* was successfully downloaded and the custom model was built on top of *ResNet50V2*.

### 3.2 Compiling the model


The following step configures the model for training.
Team picked the Adam Optimizer for its efficient updates, momentum term, and robustness to noisy gradients make it a good choice for optimizing the weights in the network during training, and its default hyperparameters make it easy to use without requiring extensive tuning.

````python
model.compile(optimizer=keras.optimizers.Adam(),
              loss=keras.losses.CategoricalCrossentropy(),
              metrics=[metrics.CategoricalAccuracy(name='accuracy'),
                       metrics.Recall(name='recall', class_id=0),
                       metrics.Precision(name='precision')])
````
